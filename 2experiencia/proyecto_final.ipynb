##**Regresión Lineal**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

#Visualizo el cuadro para seleccionar las variables
print ("Primeras filas del DataFrame")
print (df.head())
print ("\n Columnas del DataFrame")
print (df.columns)
```

```python
# Muestro las variables categoricas
df_encoded= pd.get_dummies(df)
df_encoded
```

```python
#Muesro las nuevas columnas despues del encondig
print(df_encoded.columns)
```

```python
#Seleccionamos las variables independientes (features) y la dependiente (target
feature_columns= ['hour_of_day']+[col for col in df_encoded.columns if col.startswith('coffee_name_')]
x = df_encoded[feature_columns]
y= df_encoded['money']

print(df_encoded[feature_columns])
print(df_encoded['money'])
```

```python
# Divido los datos en conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(f"\nForma de x_train: {x_train.shape}, x_test: {x_test.shape}")
```

### Modelo de Regresión Lineal
#### Regresión Lineal Simple

```python
# Creo el modelo de Regresión Líneal
modelo_regresion = LinearRegression()

# Entreno el modelo
modelo_regresion.fit(x_train, y_train)

# Realizo una predicción para las columna independiente ('money')
y_pred = modelo_regresion.predict(x_test)
print (f"Intercepto (b): {modelo_regresion.intercept_}")
print (f"Coeficientes (m) para: {'money'} {modelo_regresion.coef_}")
sales_amount= modelo_regresion.intercept_ + modelo_regresion.coef_[0] * x
```

```python
#Evaluamos el modelo con metricas
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
print("\n--- Métricas de Evaluación ---")
print(f'Error Absoluto Medio (MAE): {mae:.2f}')
print(f'Error Cuadrático Medio (MSE): {mse:.2f}')
print(f'Coeficiente de Determinación (R^2): {r2:.2f}')
```

```python
#Gráficas de dispersión actual vs predicho
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # Línea de referencia perfecta
plt.xlabel('Valores Reales (money)')
plt.ylabel('Valores Predichos (money)')
plt.title('Predicciones vs Valores Reales')
plt.show()
```

#### Regresión Lineal Multiple

```python
# Manejo las variables categóricas ('coffee_name' y 'cash_type') usando one-hot encoding
df_encoded = pd.get_dummies(df, columns=['coffee_name', 'cash_type'], drop_first=True)
```

```python
#Muestro las nuevas columnas depués del encondig
print ("Columnas despues del encoding")
print(df_encoded.columns)
```

```python
"""Selecciono las variables independientes (features) y la dependiente (target)
Incluimos 'hour_of_day', las columnas dummy de 'coffee_name' y 'payment_method'"""
coffee_dummies = [col for col in df_encoded.columns if col.startswith('coffee_name_')]
payment_dummies = [col for col in df_encoded.columns if col.startswith('payment_method_')]
feature_columns = ['hour_of_day'] + coffee_dummies + payment_dummies
X = df_encoded[feature_columns]
y = df_encoded['money']
print(f"\nForma de X: {X.shape}")
print(f"Forma de y: {y.shape}")
```

```python
#Dividimos los datos en conjuntos de entrenamiento y prueba (70% train, 30% test)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nForma de x_train: {x_train.shape}, x_test: {x_test.shape}")
```

```python
# Creo el modelo de Regresión Líneal
modelo_regresion = LinearRegression()

# Entreno el modelo
modelo_regresion.fit(x_train, y_train)

# Realizo una predicción para las tres columnas independientes ('money', 'hour_of_day)
y_pred = modelo_regresion.predict(x_test)
print (f"Intercepto (b): {modelo_regresion.intercept_}")
print (f"Coeficientes (m) para: {'money', 'hour_of_day','Weekdaysort'} {modelo_regresion.coef_}")
```

```python
# Evaluo el modelo con métricas
import numpy as np  # Importamos numpy para sqrt
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(y_test, y_pred)
print("\nResultados de la regresión lineal múltiple:")
print(f"MAE (Error Absoluto Medio): {mae:.2f}")
print(f"MSE (Error Cuadrático Medio): {mse:.2f}")
print(f"RMSE (Raíz del Error Cuadrático Medio): {rmse:.2f}")
print(f"R² (Coeficiente de determinación): {r2:.2f}")
```

```python
#Gráfico de dispersión actual vs predicho
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # Línea de referencia perfecta
plt.xlabel('Valores Reales (money)')
plt.ylabel('Valores Predichos (money)')
plt.title('Predicciones vs Valores Reales - Regresión Múltiple')
plt.show()

datos_x = np.linspace()
```
