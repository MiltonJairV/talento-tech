##1. Estadisticas para columnas númericas y categoricas

```python
import csv
import random
# Creo un nuevo achivo para trabajar y practicar lo visto.

# Definimos los encabezados de las columnas
encabezados = ['ID', 'Estudiante', 'Grado', 'Materia', 'Calificacion']

# Creamos una lista para almacenar las 20 filas de datos
datos = []

# Listas de nombres, grados y materias para generar datos aleatorios
nombres = ['Ana', 'Luis', 'Sofia', 'Pedro', 'Maria', 'Carlos', 'Laura', 'Juan', 'Elena', 'Diego', 'Lucia', 'Andres', 'Paula', 'Javier', 'Valeria', 'Daniel', 'Camila', 'Felipe', 'Isabel', 'Ricardo']
grados = ['1º Primaria', '2º Primaria', '3º Primaria', '4º Primaria', '5º Primaria', '6º Primaria']
materias = ['Matematicas', 'Ciencias', 'Historia', 'Lenguaje', 'Educacion Fisica']

# Generamos 20 filas de datos
for i in range(1, 21):
    id_estudiante = i
    nombre_estudiante = random.choice(nombres) #ramdom.choice funcion para generar datos aleatorios.
    grado_estudiante = random.choice(grados)
    materia_estudiante = random.choice(materias)
    calificacion_estudiante = random.randint(60, 100) # Calificaciones aleatorias entre 60 y 100

    fila = [id_estudiante, nombre_estudiante, grado_estudiante, materia_estudiante, calificacion_estudiante]
    datos.append(fila)

# Escribimos los datos en un archivo CSV
with open('datos_educacion.csv', 'w', newline='', encoding='utf-8') as archivo_csv:
    escritor_csv = csv.writer(archivo_csv)
    # Escribimos los encabezados
    escritor_csv.writerow(encabezados)
    # Escribimos las 20 filas de datos
    escritor_csv.writerows(datos)

print("El archivo 'datos_educacion.csv' ha sido creado exitosamente con 20 filas de datos.")
print(escritor_csv)

# Nombre del archivo que vamos a leer
nombre_archivo = 'datos_educacion.csv'

# Abrimos el archivo en modo de lectura ('r')
try:
    with open(nombre_archivo, 'r', newline='', encoding='utf-8') as archivo_csv:
        lector_csv = csv.reader(archivo_csv)

        print("Contenido del archivo CSV:")
        print("-" * 30)

        # Iteramos sobre cada fila del archivo
        for fila in lector_csv:
            # Imprimimos la fila completa
            print(fila)
except FileNotFoundError:
    print(f"Error: El archivo '{nombre_archivo}' no fue encontrado. Asegúrate de que esté en la misma carpeta.")
except Exception as e:
    print(f"Ocurrió un error: {e}")

```

## Código demo EDA Automático
Para este demo se creará un DataFrame de ejemplo con el fin de realizar la demostración, pero para dar funcionalidad, se debe reemplazar el archivo creado con el archivo a analizar en la línea:
(pd.read_csv('tu_archivo.csv')).

Elaborado con Gemini 2.5 Pro

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def eda_automatica(df):
    """
    Función para realizar un Análisis Exploratorio de Datos (EDA) automático
    y visual sobre un DataFrame de Pandas.

    Args:
        df (pd.DataFrame): El DataFrame que se desea analizar.
    """
    print("🚀 INICIANDO ANÁLISIS EXPLORATORIO DE DATOS AUTOMÁTICO 🚀")
    print("="*60)

    # --- 1. Vistazo General ---
    print("\n🧐 1. VISTAZO GENERAL DEL DATASET")
    print("-" * 40)
    print(f"Dimensiones del dataset (filas, columnas): {df.shape}")
    print("\nPrimeras 5 filas:")
    print(df.head())
    print("\nInformación general y tipos de datos:")
    # Usamos io.StringIO para capturar la salida de info() y mostrarla
    import io
    buffer = io.StringIO()
    df.info(buf=buffer)
    info_str = buffer.getvalue()
    print(info_str)

    # --- 2. Estadísticas Descriptivas y Valores Faltantes ---
    print("\n📊 2. ESTADÍSTICAS Y VALORES FALTANTES")
    print("-" * 40)
    print("Resumen estadístico para todas las columnas:")
    print(df.describe(include='all').round(2))
    print("\nConteo de valores nulos por columna:")
    print(df.isnull().sum())
    print("="*60)

    # --- 3. Análisis Gráfico de Cada Variable (Univariado) ---
    print("\n📈 3. ANÁLISIS UNIVARIADO (DISTRIBUCIÓN DE CADA COLUMNA)")
    print("-" * 40)

    for column in df.columns:
        plt.figure(figsize=(10, 4))

        # Si la columna es numérica
        if pd.api.types.is_numeric_dtype(df[column]):
            print(f"Análisis de la variable numérica: '{column}'")
            # Gráfico 1: Histograma y Densidad
            plt.subplot(1, 2, 1)
            sns.histplot(df[column], kde=True, bins=20)
            plt.title('Histograma y Densidad')

            # Gráfico 2: Box Plot para ver outliers
            plt.subplot(1, 2, 2)
            sns.boxplot(y=df[column])
            plt.title('Diagrama de Caja (Box Plot)')

        # Si la columna es categórica (o tipo 'object')
        else:
            print(f"Análisis de la variable categórica: '{column}'")
            # Si tiene demasiadas categorías, solo muestra las 10 más comunes
            if df[column].nunique() > 10:
                top_10 = df[column].value_counts().nlargest(10)
                sns.barplot(x=top_10.index, y=top_10.values)
                plt.title(f'Top 10 Categorías más Frecuentes')
            else:
                sns.countplot(y=df[column], order=df[column].value_counts().index)
                plt.title('Conteo de Frecuencias')

        plt.suptitle(f'Análisis de la Columna: {column}', fontsize=16)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

    print("="*60)

    # --- 4. Mapa de Calor de Correlaciones (Bivariado) ---
    # Seleccionamos solo las columnas numéricas para la matriz de correlación
    df_numeric = df.select_dtypes(include=np.number)

    if not df_numeric.empty:
        print("\n🔥 4. MAPA DE CALOR DE CORRELACIONES (VARIABLES NUMÉRICAS)")
        print("-" * 40)

        corr_matrix = df_numeric.corr()

        plt.figure(figsize=(10, 7))
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
        plt.title('Matriz de Correlación', fontsize=16)
        plt.show()
    else:
        print("\n🔥 4. MAPA DE CALOR DE CORRELACIONES: No hay suficientes variables numéricas para generar el mapa.")

    print("\n✅ ANÁLISIS EXPLORATORIO FINALIZADO.")
    print("="*60)

# --- CREACIÓN DE UN DATASET DE EJEMPLO PARA EL DEMO ---
# En tu caso, reemplazarías esta sección por: df = pd.read_csv('tu_archivo.csv')
data_ejemplo = {
    'edad': [25, 30, np.nan, 35, 28, 45, 22, 38, 30, 25],
    'salario_anual': [50000, 62000, 45000, 75000, 52000, 95000, 48000, 82000, 62000, 51000],
    'ciudad': ['Bogotá', 'Medellín', 'Bogotá', 'Cali', 'Medellín', 'Cali', 'Bogotá', 'Medellín', 'Bogotá', 'Cali'],
    'genero': ['M', 'F', 'F', 'M', 'F', 'M', 'M', 'F', 'M', np.nan],
    'satisfaccion': [4, 5, 3, 5, 4, 5, 2, 4, 3, 4]
}
mi_dataframe = pd.DataFrame(data_ejemplo)

# --- EJECUTAR LA FUNCIÓN DE EDA AUTOMÁTICO ---
eda_automatica(mi_dataframe)
```
