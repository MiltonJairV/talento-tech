```python
import pandas as pd
x=[1,2,2,3,4,4,5,6]
y=[2,3,4,4,4,6,5,7]

d = {"y":y, "x":x}
df  = pd.DataFrame(d)
df
```

```python
# Creación de la gráfica
import plotly.express as px
import matplotlib.pyplot as plt

plt.scatter(df['x'], df['y'])
```

```python
# Se crean las etiquetas en el gráfico
plt.title('Relación entre X y Y') # Titulo
plt.xlabel('Variable Independiente (x)') # Etiqueta el eje X
plt.ylabel('Variable Dependiente (y)') # Etiquta para el eje Y
plt.grid(True) # Añado cuadricula
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
modelo =LinearRegression()
modelo.fit(df[['x']], df['y'])
```

```python
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print(f"La ecuación de la linea es: Y={pendiente:.2f}x + {intercepto:.2f}")
print ("-----------------------------------------")
print (f"La pendiente es (m): {pendiente}")
print (f"El intercepto es (: {intercepto:}")
```

```python
# Usar el modelo para predecir los valores
y_predichos = modelo.predict(df[['x']])

#Dibujar los datos originales y la línea del modelo
plt.scatter(df['x'], df['y'], color='blue', label='Datos originales')

#Dibuja la línea de regresión en rojo
plt.plot(df['x'], y_predichos, color='red', label='Regresión lineal')

# Se crean las etiquetas en el gráfico
plt.title('Relación entre X y Y') # Titulo
plt.xlabel('Variable Independiente (x)') # Etiqueta el eje X
plt.ylabel('Variable Dependiente (y)') # Etiquta para el eje Y
plt.grid(True) # Añado cuadricula
plt.show()
```
