### Modelo de Regresi√≥n Lineal
#### Regresi√≥n Lineal Simple

```python
# Selecciono la variable independiente (feature) y la variable dependiente (target) para la regresi√≥n lineal simple
# Uso 'hour_of_day'como caracter√≠stica √∫nica para demostrar un coeficiente √∫nico.
x = df[['hour_of_day']]  # selecciono solo la columna 'hour_of_day' como un dataframe
y = df['money']          # Selecciono la columna 'money' como variable dependiente

# Divido los datos en conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Create the Simple Linear Regression model
modelo_regresion = LinearRegression()

# Entreno el modelo
modelo_regresion.fit(x_train, y_train)

# Hago la predicci√≥n
y_pred = modelo_regresion.predict(x_test)

print(f"Intercept (b): {modelo_regresion.intercept_}")

# # Imprime el coeficiente √∫nico para 'hora_del_d√≠a'
print(f"Coefficient (m) for 'hour_of_day': {modelo_regresion.coef_[0]}")
```

```python
#Evaluamos el modelo con metricas
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
print("\n--- M√©tricas de Evaluaci√≥n ---")
print(f'Error Absoluto Medio (MAE): {mae:.2f}')
print(f'Error Cuadr√°tico Medio (MSE): {mse:.2f}')
print(f'Coeficiente de Determinaci√≥n (R^2): {r2:.2f}')
```

```python
#Gr√°ficas de dispersi√≥n actual vs predicho
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.3)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # L√≠nea de referencia perfecta
plt.xlabel('Valores Reales (money)')
plt.ylabel('Valores Predichos (money)')
plt.title('Predicciones vs Valores Reales')
plt.show()
```

#### Regresi√≥n Lineal Multiple

```python
# Manejo las variables categ√≥ricas ('coffee_name' y 'cash_type') usando one-hot encoding
df_encoded = pd.get_dummies(df, columns=['coffee_name', 'cash_type'], drop_first=True)
```

```python
#Muestro las nuevas columnas depu√©s del encondig
print ("Columnas despues del encoding")
print(df_encoded.columns)
```

```python
"""Selecciono las variables independientes (features) y la dependiente (target)
Incluimos 'hour_of_day', las columnas dummy de 'coffee_name' y 'payment_method'"""
coffee_dummies = [col for col in df_encoded.columns if col.startswith('coffee_name_')]
payment_dummies = [col for col in df_encoded.columns if col.startswith('payment_method_')]
feature_columns = ['hour_of_day'] + coffee_dummies + payment_dummies
X = df_encoded[feature_columns]
y = df_encoded['money']
print(f"\nForma de X: {X.shape}")
print(f"Forma de y: {y.shape}")
```

```python
#Dividimos los datos en conjuntos de entrenamiento y prueba (70% train, 30% test)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"\nForma de x_train: {x_train.shape}, x_test: {x_test.shape}")
```

```python
# Creo el modelo de Regresi√≥n M√∫ltiple
modelo_regresion = LinearRegression()

# Entreno el modelo
modelo_regresion.fit(x_train, y_train)

# Realizo una predicci√≥n para las tres columnas independientes ('money', 'hour_of_day)
y_pred = modelo_regresion.predict(x_test)
print (f"Intercepto (b): {modelo_regresion.intercept_}")
print (f"Coeficientes (m) para: {'money', 'hour_of_day','Weekdaysort'} {modelo_regresion.coef_}")
```

```python
# Evaluo el modelo con m√©tricas
import numpy as np  # Importamos numpy para sqrt
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(y_test, y_pred)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print("\nResultados de la regresi√≥n lineal m√∫ltiple:")
print(f"MAE (Error Absoluto Medio): {mae:.2f}")
print(f"MSE (Error Cuadr√°tico Medio): {mse:.2f}")
print(f"RMSE (Ra√≠z del Error Cuadr√°tico Medio): {rmse:.2f}")
print(f"R¬≤ (Coeficiente de determinaci√≥n): {r2:.2f}")
print(f"MAPE (Error Porcentual Absoluto Medio): {mape:.2f}%")
```

```python
#Gr√°fico de dispersi√≥n actual vs predicho
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)  # L√≠nea de referencia perfecta
plt.xlabel('Valores Reales (money)')
plt.ylabel('Valores Predichos (money)')
plt.title('Predicciones vs Valores Reales - Regresi√≥n M√∫ltiple')
plt.show()


```

```python
# Print the coefficients of the model
print("\nCoeficientes del modelo:")
# Since the model was trained with 'hour_of_day' as the only feature,
# print the coefficient for 'hour_of_day' and the intercept.
print(f"hour_of_day: {modelo_regresion.coef_[0]:.2f}")
print(f"Intercepto: {modelo_regresion.intercept_:.2f}")
```

```python
# Gr√°fico el Histograma de los valores de prueba (reales)
plt.figure(figsize=(12, 6))
sns.histplot(y_test, kde=True, bins=30, color='blue', label='Valores Reales (y_test)')
plt.title('Distribuci√≥n de los Valores Reales de "money"')
plt.xlabel('Valores Reales')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
# Gr√°fico el histograma de los valores predichos
plt.figure(figsize=(12, 6))
sns.histplot(y_pred, kde=True, bins=30, color='green', label='Valores Predichos (y_pred)')
plt.title('Distribuci√≥n de los Valores Predichos de "money"')
plt.xlabel('Valores Predichos')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
# Gr√°fico del Histogr√°ma comparativo
plt.figure(figsize=(12, 6))
sns.histplot(y_test, kde=True, bins=30, color='blue', label='Valores Reales', alpha=0.5)
sns.histplot(y_pred, kde=True, bins=30, color='orange', label='Valores Predichos', alpha=0.5)
plt.title('Comparaci√≥n de la Distribuci√≥n: Reales vs. Predichos')
plt.xlabel('Valores de "money"')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Asumiendo que las variables x_test y y_test ya han sido creadas
# con el c√≥digo de regresi√≥n lineal simple que proporcionaste.

# Se entrena el modelo de regresi√≥n lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(x_train, y_train)

# Se realizan las predicciones para el conjunto de prueba
y_pred = modelo_regresion.predict(x_test)

# Se calculan los errores (residuos)
errores = y_test - y_pred

# Se calcula la media y la desviaci√≥n est√°ndar de los errores
media_errores = np.mean(errores)
desviacion_estandar_errores = np.std(errores)

print(f"Media de los errores: {media_errores:.2f}")
print(f"Desviaci√≥n est√°ndar de los errores: {desviacion_estandar_errores:.2f}")

# Se crea el histograma de los errores
plt.figure(figsize=(10, 6))
sns.histplot(errores, bins=30, kde=True, color='purple')
plt.title('Distribuci√≥n de los Errores (Residuos) del Modelo')
plt.xlabel('Errores (Valores Reales - Valores Predichos)')
plt.ylabel('Frecuencia')
plt.axvline(x=0, color='red', linestyle='--', label='Error = 0') # L√≠nea en el 0
plt.legend()
plt.show()
```
